# Polynomial Regression

```python
import seaborn as sns
import numpy as np

from matplotlib import pyplot as plt
```

```python
np.random.seed(42)

m = 100
X = 6 * np.random.rand(m, 1) - 3
# Using random.randn() for the Normal Distribution
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)
```

```python
sns.scatterplot(x=X[:, 0], y=y[:, 0], color='blue', marker='.')

plt.axis((-3.0, 3.0, 0.0, 10.0))
plt.xlabel('x1')
plt.ylabel('y')

plt.plot()
```

### Using Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

print('x[0]:', X[0])
print('X_poly[0]:', X_poly[0])
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

lin_reg.intercept_, lin_reg.coef_
```

```python
y_pred = lin_reg.predict(X_poly)

sns.scatterplot(x=X[:, 0], y=y[:, 0], color='blue', marker='.')
sns.lineplot(x=X[:, 0], y=y_pred[:, 0], color='red', label='Predictions')

plt.axis((-3, 3, 0, 10))
plt.legend(loc='upper left')
plt.figure(1)
plt.show()
```

```python
print("Polynomial Regression")
print("y_pred intercept", lin_reg.intercept_)
print("y_pred coeff", lin_reg.coef_)
```
